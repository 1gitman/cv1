```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import plotly.express as px # For interactive plots if needed, though static is fine for this context
import plotly.graph_objects as go

# Set style for better visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# --- Data Loading and Initial Exploration ---
# Load all datasets
try:
    blocks = pd.read_csv("blocks.csv", index_col="Unnamed: 0")
    scoring = pd.read_csv("scoring.csv", index_col="Unnamed: 0")
    rebounds = pd.read_csv("rebounds.csv", index_col="Unnamed: 0")
    assists_turnovers = pd.read_csv("assists-turnovers.csv", index_col="Unnamed: 0")
    steals = pd.read_csv("steals.csv", index_col="Unnamed: 0")
    fouls_minutes = pd.read_csv("fouls-minutes.csv", index_col="Unnamed: 0")
    print("All dataframes loaded successfully from CSVs.")
except FileNotFoundError as e:
    print(f"Error loading a CSV file: {e}. Creating dummy dataframes for demonstration.")
    # Create dummy dataframes with consistent column names and types for demonstration
    num_players = 250
    players = [f'Player_{i}' for i in range(num_players)]
    positions = np.random.choice(['PG', 'SG', 'SF', 'PF', 'C'], num_players)
    teams = np.random.choice([f'Team_{chr(65+i)}' for i in range(10)], num_players)

    # Scoring DataFrame
    scoring_data = {
        'Player': players,
        'Position': positions,
        'Team': teams,
        'GP  Games played': np.random.randint(10, 82, num_players),
        'GS  Games started': np.random.randint(0, 82, num_players),
        'MPG  Minutes Per Game': [f"{np.random.randint(10, 40)}:{np.random.randint(0, 59):02d}" for _ in range(num_players)], # String format for MPG
        'PPG  Points Per Game': np.random.uniform(5.0, 30.0, num_players).round(1),
        'FGM  Field Goals Made': np.random.randint(1, 15, num_players),
        'FGA  Field Goals Attempted': np.random.randint(5, 30, num_players),
        'FG%  Field Goal Percentage': [f"{np.random.uniform(0.3, 0.6):.3f}%" for _ in range(num_players)], # String format with %
        '3FGM  Three-Point Field Goals Made': np.random.randint(0, 5, num_players),
        '3FGA  Three-Point Field Goals Attempted': np.random.randint(0, 10, num_players),
        '3FG%  Three-Point Field Goal Percentage': [f"{np.random.uniform(0.2, 0.4):.3f}%" for _ in range(num_players)], # String format with %
        'FTM  Free Throws Made': np.random.randint(0, 10, num_players),
        'FTA  Free Throws Attempted': np.random.randint(0, 12, num_players),
        'FT%  Free Throw Percentage': [f"{np.random.uniform(0.7, 0.9):.3f}%" for _ in range(num_players)], # String format with %
    }
    scoring = pd.DataFrame(scoring_data)

    # Other DataFrames with relevant columns for merging
    rebounds = pd.DataFrame({
        'Player': players,
        'RPG  Rebounds Per Game': np.random.uniform(1.0, 15.0, num_players).round(1)
    })
    assists_turnovers = pd.DataFrame({
        'Player': players,
        'APG  Assists Per Game': np.random.uniform(0.5, 10.0, num_players).round(1),
        'TOPG  Turnovers Per Game': np.random.uniform(0.1, 5.0, num_players).round(1)
    })
    steals = pd.DataFrame({
        'Player': players,
        'SPG  Steals Per Game': np.random.uniform(0.1, 3.0, num_players).round(1)
    })
    blocks = pd.DataFrame({
        'Player': players,
        'BPG  Blocks Per Game': np.random.uniform(0.0, 5.0, num_players).round(1)
    })
    # For fouls_minutes, include MPG if it's expected in the actual CSV to cause conflict
    fouls_minutes = pd.DataFrame({
        'Player': players,
        'FPG  Fouls Per Game': np.random.uniform(1.0, 5.0, num_players).round(1),
        'MPG  Minutes Per Game': [f"{np.random.randint(10, 40)}:{np.random.randint(0, 59):02d}" for _ in range(num_players)] # Added to simulate potential conflict
    })

print("Scoring data shape:", scoring.shape)
print("Rebounds data shape:", rebounds.shape)
print("Assists-Turnovers data shape:", assists_turnovers.shape)
print("Steals data shape:", steals.shape)
print("Blocks data shape:", blocks.shape)
print("Fouls-Minutes data shape:", fouls_minutes.shape)

# --- Data Cleaning and Preparation ---
# Clean percentage columns in scoring data
def clean_percentage(perc):
    """Converts a percentage string (e.g., '34.5%') to a float (0.345).
    Handles non-string inputs by converting to float directly if possible.
    """
    if isinstance(perc, str):
        try:
            return float(perc.strip('%')) / 100
        except ValueError:
            return np.nan # Return NaN if conversion fails
    return float(perc) # Already a number, just ensure float type

scoring['FG%  Field Goal Percentage'] = scoring['FG%  Field Goal Percentage'].apply(clean_percentage)
scoring['3FG%  Three-Point Field Goal Percentage'] = scoring['3FG%  Three-Point Field Goal Percentage'].apply(clean_percentage)
scoring['FT%  Free Throw Percentage'] = scoring['FT%  Free Throw Percentage'].apply(clean_percentage)

# Merge all datasets on 'Player'
# Start with scoring as the base, then left merge others
merged_df = scoring.copy()
dfs_to_merge = [rebounds, assists_turnovers, steals, blocks, fouls_minutes]

for df in dfs_to_merge:
    # Identify columns in the current df that are also in merged_df (excluding 'Player')
    # and are not intended to be duplicated (like 'Position', 'Team', 'GP', 'GS', 'MPG')
    common_cols_to_drop = [
        col for col in ['Position', 'Team', 'GP  Games played', 'GS  Games started', 'MPG  Minutes Per Game']
        if col in df.columns and col != 'Player'
    ]

    # Drop these common columns from the current df before merging to avoid suffix issues
    df_to_merge = df.drop(columns=common_cols_to_drop, errors='ignore') # Use errors='ignore' in case column doesn't exist

    # Use 'Player' as the common key for merging
    merged_df = pd.merge(merged_df, df_to_merge, on='Player', how='left')

# Check for missing values after merging
print("\nMissing values in merged dataset (before filling):")
print(merged_df.isnull().sum())

# Fill missing values with 0 (assuming missing means 0 in basketball stats)
# This is crucial for numerical operations later.
merged_df.fillna(0, inplace=True)
print("\nMissing values after filling with 0:")
print(merged_df.isnull().sum())

# Convert minutes to numeric (handling the "MM:SS" format)
def convert_mpg_to_float(mpg_str):
    """Converts 'MM:SS' string to total minutes as a float."""
    if isinstance(mpg_str, str) and ':' in mpg_str:
        try:
            minutes, seconds = map(float, mpg_str.split(':'))
            return minutes + seconds / 60
        except ValueError:
            return np.nan # Handle cases where conversion might fail
    return float(mpg_str) # Assume it's already a number if not a string or no ':'

merged_df['MPG  Minutes Per Game'] = merged_df['MPG  Minutes Per Game'].apply(convert_mpg_to_float)
# Fill any NaNs that might result from conversion errors after this step
merged_df['MPG  Minutes Per Game'].fillna(0, inplace=True)


# Create efficiency metrics
# Ensure all columns used here exist in merged_df
merged_df['Efficiency'] = (merged_df['PPG  Points Per Game'] +
                          merged_df['RPG  Rebounds Per Game'] +
                          merged_df['APG  Assists Per Game'] +
                          merged_df['SPG  Steals Per Game'] +
                          merged_df['BPG  Blocks Per Game'] -
                          merged_df['TOPG  Turnovers Per Game'])

# True Shooting % calculation - ensure denominator is not zero
merged_df['True Shooting %'] = merged_df['PPG  Points Per Game'] / (
    2 * (merged_df['FGA  Field Goals Attempted'] + 0.44 * merged_df['FTA  Free Throws Attempted']))
# Handle division by zero (results in inf), replace with 0 or NaN as appropriate
merged_df['True Shooting %'] = merged_df['True Shooting %'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Calculate Assist-to-Turnover ratio (moved here to be before usage)
merged_df['AST/TO'] = merged_df['APG  Assists Per Game'] / merged_df['TOPG  Turnovers Per Game']
merged_df['AST/TO'] = merged_df['AST/TO'].replace([np.inf, -np.inf], np.nan).fillna(0) # Replace inf with NaN then fill 0

# Player Impact Estimate (PIE) - simplified version (moved here to be before usage)
# Ensure all columns are present before calculation
merged_df['PIE'] = (merged_df['PPG  Points Per Game'] +
                   merged_df['RPG  Rebounds Per Game'] +
                   merged_df['APG  Assists Per Game'] +
                   merged_df['SPG  Steals Per Game'] +
                   merged_df['BPG  Blocks Per Game'] -
                   merged_df['TOPG  Turnovers Per Game'] -
                   merged_df['FPG  Fouls Per Game'])

# Usage Rate (simplified) (moved here to be before usage)
# Handle potential division by zero for MPG
merged_df['Usage'] = (merged_df['FGA  Field Goals Attempted'] +
                     0.44 * merged_df['FTA  Free Throws Attempted'] +
                     merged_df['TOPG  Turnovers Per Game']) / (
    merged_df['MPG  Minutes Per Game'].replace(0, np.nan) / 48) # Replace 0 with NaN to avoid div by zero
merged_df['Usage'].fillna(0, inplace=True) # Fill NaNs resulting from div by zero with 0


print("\nMerged DataFrame Info after cleaning and feature engineering:")
merged_df.info()
print("\nSample of Merged DataFrame:")
display(merged_df.head())

# --- Exploratory Data Analysis ---
print("\n--- Exploratory Data Analysis ---")

# 1. Distribution of Key Metrics
print("\n1. Distribution of Key Metrics")
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
metrics = ['PPG  Points Per Game', 'RPG  Rebounds Per Game', 'APG  Assists Per Game',
           'SPG  Steals Per Game', 'BPG  Blocks Per Game', 'Efficiency']

for i, metric in enumerate(metrics):
    row, col = i // 3, i % 3
    sns.histplot(merged_df[metric], ax=axes[row, col], kde=True)
    axes[row, col].set_title(f'Distribution of {metric.split("  ")[0]}')
    axes[row, col].set_xlabel('')

plt.tight_layout()
plt.suptitle('Distribution of Key Player Metrics', y=1.02, fontsize=16)
plt.show()

# 2. Efficiency by Position
print("\n2. Efficiency by Position")
plt.figure(figsize=(12, 6))
sns.boxplot(x='Position', y='Efficiency', data=merged_df, order=['PG', 'SG', 'SF', 'PF', 'C'], palette='viridis')
plt.title('Player Efficiency by Position', fontsize=14)
plt.xlabel('Position', fontsize=12)
plt.ylabel('Efficiency Score', fontsize=12)
plt.show()

# Top 10 most efficient players
top_10_efficient = merged_df.sort_values('Efficiency', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Efficiency', y='Player', data=top_10_efficient, hue='Position', dodge=False, palette='coolwarm')
plt.title('Top 10 Most Efficient Players', fontsize=14)
plt.xlabel('Efficiency Score', fontsize=12)
plt.ylabel('Player Name', fontsize=12)
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3. Correlation Heatmap
print("\n3. Correlation Heatmap")
# Select numeric columns for correlation
numeric_cols = merged_df.select_dtypes(include=[np.number]).columns
corr_df = merged_df[numeric_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_df, annot=True, fmt='.2f', cmap='coolwarm', center=0, linewidths=.5, linecolor='black')
plt.title('Correlation Heatmap of Player Metrics', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# 4. Shooting Efficiency Analysis
print("\n4. Shooting Efficiency Analysis")
fig, axes = plt.subplots(1, 3, figsize=(20, 6)) # Increased figure width

# FG% vs PPG
sns.scatterplot(x='FG%  Field Goal Percentage', y='PPG  Points Per Game',
                data=merged_df, hue='Position', ax=axes[0], palette='deep', alpha=0.7)
axes[0].set_title('Field Goal % vs Points Per Game', fontsize=14)
axes[0].set_xlabel('Field Goal Percentage', fontsize=12)
axes[0].set_ylabel('Points Per Game', fontsize=12)
axes[0].legend(title='Position')

# 3FG% vs 3FGM
sns.scatterplot(x='3FG%  Three-Point Field Goal Percentage',
                y='3FGM  Three-Point Field Goals Made',
                data=merged_df, hue='Position', ax=axes[1], palette='deep', alpha=0.7)
axes[1].set_title('3PT % vs 3PT Made', fontsize=14)
axes[1].set_xlabel('Three-Point Field Goal Percentage', fontsize=12)
axes[1].set_ylabel('Three-Point Field Goals Made', fontsize=12)
axes[1].legend(title='Position')

# FT% vs FTM
sns.scatterplot(x='FT%  Free Throw Percentage',
                y='FTM  Free Throws Made',
                data=merged_df, hue='Position', ax=axes[2], palette='deep', alpha=0.7)
axes[2].set_title('FT % vs FT Made', fontsize=14)
axes[2].set_xlabel('Free Throw Percentage', fontsize=12)
axes[2].set_ylabel('Free Throws Made', fontsize=12)
axes[2].legend(title='Position')

plt.tight_layout()
plt.show()

# 5. Minutes Played vs Efficiency
print("\n5. Minutes Played vs Efficiency")
plt.figure(figsize=(12, 7))
sns.scatterplot(x='MPG  Minutes Per Game', y='Efficiency',
                data=merged_df, hue='Position', size='PPG  Points Per Game',
                sizes=(50, 800), alpha=0.7, palette='tab10') # Adjusted sizes for better visibility
plt.title('Minutes Played vs Efficiency (Size by PPG)', fontsize=14)
plt.xlabel('Minutes Per Game', fontsize=12)
plt.ylabel('Efficiency Score', fontsize=12)
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 6. Assist-to-Turnover Ratio Analysis (This calculation is now moved earlier)
print("\n6. Assist-to-Turnover Ratio Analysis")
plt.figure(figsize=(12, 6))
sns.boxplot(x='Position', y='AST/TO', data=merged_df, order=['PG', 'SG', 'SF', 'PF', 'C'], palette='pastel')
plt.title('Assist-to-Turnover Ratio by Position', fontsize=14)
plt.xlabel('Position', fontsize=12)
plt.ylabel('Assist-to-Turnover Ratio', fontsize=12)
plt.ylim(0, merged_df['AST/TO'].quantile(0.95) * 1.1) # Limit y-axis based on data for better visualization
plt.show()

# 7. Team Performance Analysis
print("\n7. Team Performance Analysis")
team_stats = merged_df.groupby('Team').agg(
    PPG_mean=('PPG  Points Per Game', 'mean'),
    RPG_mean=('RPG  Rebounds Per Game', 'mean'),
    APG_mean=('APG  Assists Per Game', 'mean'),
    SPG_mean=('SPG  Steals Per Game', 'mean'),
    BPG_mean=('BPG  Blocks Per Game', 'mean'),
    Efficiency_mean=('Efficiency', 'mean'),
    Player_Count=('Player', 'count')
).sort_values('Efficiency_mean', ascending=False)

# Top 10 teams by average player efficiency
plt.figure(figsize=(12, 7))
sns.barplot(x='Efficiency_mean', y=team_stats.head(10).index, data=team_stats.head(10), palette='rocket')
plt.title('Top 10 Teams by Average Player Efficiency', fontsize=14)
plt.xlabel('Average Efficiency', fontsize=12)
plt.ylabel('Team', fontsize=12)
plt.tight_layout()
plt.show()

# 8. Position-wise Performance Comparison
print("\n8. Position-wise Performance Comparison")
position_stats = merged_df.groupby('Position').agg(
    PPG_mean=('PPG  Points Per Game', 'mean'),
    RPG_mean=('RPG  Rebounds Per Game', 'mean'),
    APG_mean=('APG  Assists Per Game', 'mean'),
    SPG_mean=('SPG  Steals Per Game', 'mean'),
    BPG_mean=('BPG  Blocks Per Game', 'mean'),
    Efficiency_mean=('Efficiency', 'mean')
).sort_values('Efficiency_mean', ascending=False)

# Plot position stats
position_stats.plot(kind='bar', figsize=(14, 7), colormap='Paired')
plt.title('Average Performance Metrics by Position', fontsize=14)
plt.ylabel('Average Value', fontsize=12)
plt.xlabel('Position', fontsize=12)
plt.xticks(rotation=0)
plt.legend(loc='upper right', title='Metric')
plt.tight_layout()
plt.show()

# --- Enhanced Player Clustering Analysis ---
print("\n--- Enhanced Player Clustering Analysis ---")

# Select features for clustering
cluster_features_cols = ['PPG  Points Per Game', 'APG  Assists Per Game',
                         'RPG  Rebounds Per Game', 'SPG  Steals Per Game',
                         'BPG  Blocks Per Game', 'FG%  Field Goal Percentage',
                         'True Shooting %', 'Efficiency', 'AST/TO', 'Usage', 'PIE'] # Added PIE

# Ensure all selected columns exist in the DataFrame
# Filter out any columns that might not exist in the dummy data or actual data
cluster_features_cols = [col for col in cluster_features_cols if col in merged_df.columns]
cluster_features = merged_df[cluster_features_cols]

# Scale the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(cluster_features)
scaled_features_df = pd.DataFrame(scaled_features, columns=cluster_features_cols)


# 9. Elbow Method to Determine Optimal Number of Clusters (K)
print("\n9. Elbow Method for Optimal K")
wcss = []
for i in range(1, 11): # Test K from 1 to 10
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K', fontsize=14)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Within-Cluster Sum of Squares (WCSS)', fontsize=12)
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
print("Look for the 'elbow' point in the graph to determine the optimal number of clusters.")

# 10. Silhouette Score to Evaluate Cluster Quality
print("\n10. Silhouette Score for Cluster Quality")
silhouette_scores = []
# Silhouette score requires at least 2 clusters
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(scaled_features)
    score = silhouette_score(scaled_features, cluster_labels)
    silhouette_scores.append(score)
    print(f"For K={i}, Silhouette Score: {score:.3f}")

plt.figure(figsize=(10, 6))
plt.bar(range(2, 11), silhouette_scores, color='skyblue')
plt.title('Silhouette Scores for Different K Values', fontsize=14)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Silhouette Score', fontsize=12)
plt.xticks(range(2, 11))
plt.grid(axis='y')
plt.show()
print("Higher Silhouette Scores indicate better-defined and separated clusters. Choose K based on a balance between Elbow Method and Silhouette Score.")


# Re-run K-means with the chosen optimal K (e.g., let's assume K=4 for demonstration)
# User should choose based on Elbow/Silhouette plots
optimal_k = 4 # Example: User might choose 4 based on the plots
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
merged_df['Cluster'] = kmeans.fit_predict(scaled_features)

# 11. Visualize Clusters with Pair Plot
print("\n11. Pair Plot of Key Metrics by Cluster")
# Select a subset of key features for the pair plot to keep it readable
pair_plot_features = ['PPG  Points Per Game', 'APG  Assists Per Game', 'RPG  Rebounds Per Game', 'Efficiency']
sns.pairplot(merged_df, vars=pair_plot_features, hue='Cluster', palette='viridis', diag_kind='kde')
plt.suptitle('Pair Plot of Key Metrics by Player Cluster', y=1.02, fontsize=16)
plt.show()
print("This plot helps visualize how different clusters separate across various metric pairs.")

# 12. Radar Chart for Cluster Profiles
print("\n12. Radar Chart of Cluster Profiles")
# Calculate mean of scaled features for each cluster
cluster_means_scaled = scaled_features_df.copy()
cluster_means_scaled['Cluster'] = merged_df['Cluster'] # Add cluster labels to scaled data
cluster_profiles = cluster_means_scaled.groupby('Cluster')[cluster_features_cols].mean().reset_index()

# Create radar chart for each cluster
fig = go.Figure()

for i in range(optimal_k):
    cluster_data = cluster_profiles[cluster_profiles['Cluster'] == i]
    fig.add_trace(go.Scatterpolar(
        r=cluster_data[cluster_features_cols].values[0],
        theta=cluster_features_cols,
        fill='toself',
        name=f'Cluster {i}'
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[scaled_features.min(), scaled_features.max()] # Set range based on overall scaled data
        )),
    showlegend=True,
    title='Radar Chart of Player Cluster Profiles (Scaled Metrics)',
    title_font_size=16
)
fig.show()
print("Radar charts visually represent the average profile of players within each cluster across selected metrics, making it easy to compare cluster strengths and weaknesses.")


# Analyze cluster characteristics (updated to include new metrics)
cluster_analysis = merged_df.groupby('Cluster').agg(
    PPG_mean=('PPG  Points Per Game', 'mean'),
    APG_mean=('APG  Assists Per Game', 'mean'),
    RPG_mean=('RPG  Rebounds Per Game', 'mean'),
    SPG_mean=('SPG  Steals Per Game', 'mean'),
    BPG_mean=('BPG  Blocks Per Game', 'mean'),
    FG_Perc_mean=('FG%  Field Goal Percentage', 'mean'),
    True_Shooting_Perc_mean=('True Shooting %', 'mean'),
    Efficiency_mean=('Efficiency', 'mean'),
    AST_TO_mean=('AST/TO', 'mean'),
    Usage_mean=('Usage', 'mean'),
    PIE_mean=('PIE', 'mean'), # Added PIE_mean
    Most_Common_Position=('Position', lambda x: x.mode()[0]), # Get the most frequent position
    Player_Count=('Player', 'count')
).rename(columns={'Player_Count': 'Count'}).round(2) # Round numerical columns for display

print("\nUpdated Cluster Characteristics:")
display(cluster_analysis)
print("This table provides a quantitative summary of each cluster's average performance across all key metrics.")


# 13. Top Players per Cluster (Example for one cluster)
print("\n13. Top Players within Each Cluster (Example for Cluster 0)")
# You can loop through all clusters if needed
for cluster_id in sorted(merged_df['Cluster'].unique()):
    print(f"\nTop 5 Players in Cluster {cluster_id} by Efficiency:")
    top_players_in_cluster = merged_df[merged_df['Cluster'] == cluster_id].sort_values(
        'Efficiency', ascending=False).head(5)
    display(top_players_in_cluster[['Player', 'Position', 'Team', 'Efficiency', 'PPG  Points Per Game', 'Usage']])

# 14. Distribution of Positions within Clusters
print("\n14. Distribution of Positions within Clusters")
cluster_position_distribution = merged_df.groupby(['Cluster', 'Position']).size().unstack(fill_value=0)
cluster_position_distribution_norm = cluster_position_distribution.apply(lambda x: x / x.sum(), axis=1)

cluster_position_distribution_norm.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')
plt.title('Distribution of Player Positions within Each Cluster', fontsize=14)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Proportion of Players', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
print("This stacked bar chart shows the positional makeup of each cluster, helping to understand if clusters are dominated by certain positions.")


# 15. Team Representation within Clusters
print("\n15. Team Representation within Clusters")
# Group by cluster and team, then count players
cluster_team_distribution = merged_df.groupby(['Cluster', 'Team']).size().unstack(fill_value=0)

# Display the table (can be large, consider displaying head/tail or specific clusters)
print("Count of Players from Each Team within Each Cluster:")
display(cluster_team_distribution.head()) # Displaying head as it can be a large table

# You could also visualize this if there are not too many teams/clusters
# For visualization, it might be better to pick top N teams or specific clusters
# Example: Heatmap of player counts per team per cluster (if number of teams/clusters is manageable)
if cluster_team_distribution.shape[0] < 10 and cluster_team_distribution.shape[1] < 20: # Heuristic to avoid too large plot
    plt.figure(figsize=(15, 8))
    sns.heatmap(cluster_team_distribution, annot=True, fmt='d', cmap='Blues')
    plt.title('Player Count per Team per Cluster', fontsize=14)
    plt.xlabel('Team', fontsize=12)
    plt.ylabel('Cluster', fontsize=12)
    plt.tight_layout()
    plt.show()
else:
    print("\nHeatmap for Team Representation within Clusters not generated due to large number of teams/clusters for better readability.")
print("This analysis helps identify if certain clusters are over-represented by players from specific teams.")
```
